## Multiple Integration with Scipy

* **dblquad**: can be used to evaluate $\displaystyle\int_a^b\int_c^df(x,y)dxdy$
* **tplquad**: can be used to evaluate $\displaystyle\int_a^b\int_c^d\int_e^ff(x,y,)dxdydz$
* **nquad**: can be used to evaluate $\displaystyle\int\dots\int_Df(\mathbf{x})d\mathbf{x}$ 

These functions are wrappers around the single-variable quadrature function **quad**, which is called repeatedly along each dimension of the multiple integral

```python
#import modules
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

<font color=blue>**Example**:
    evaluate
    $$
    \int_{-1}^1\int_{-1}^1   \sin(2\pi x)\sin(2\pi y)dxdy =  0
    $$
    
</font>




```python
f = lambda x,y: np.sin(2.*np.pi*x)* np.sin(2.*np.pi*y)
```

```python
val,err=integrate.dblquad(f,-1,1,-1,1)
print('val=%e\nerr=%e'%(val,err))
```

We can also compute integrals of the following form:
$$
\displaystyle\int_a^b\int_{g(x)}^{h(x)}f(x,y)dxdy
$$

<font color=blue>**Example**:
    evaluate
    $$
    \int_{0}^1\int_{g(x)}^{h(x)} e^{-x^2-y^2} dxdy 
    $$
</font>

We consider two cases:
1. $g(x)=0$ and $h(x)=1$, i.e., $\displaystyle
    \int_{0}^1\int_{0}^{1} e^{-x^2-y^2} dxdy 
    $
2.   $g(x)=x-1$ and $h(x)=1-x$, i.e., $\displaystyle
    \int_{0}^1\int_{x-1}^{1-x} e^{-x^2-y^2} dydx
    $


First, let's see what we are doing here

```python
# integrand
f = lambda x,y: np.exp(-x**2-y**2)

x=np.linspace(-1.,1.,100)
y=x
X,Y=np.meshgrid(x,y)  # square grid [-1,1]x[-1,1]

# integral limit along x dimension
a=0.
b=1. 
# case 1
g1 =  lambda x: 0.
h1 =  lambda x: 1.
val1,err1=integrate.dblquad(f,a,b,g1,h1)

# case 2
g2 =  lambda x: x-1.
h2 =  lambda x: 1.-x
val2,err2=integrate.dblquad(f,a,b,g2,h2)




```

```python

xint=np.linspace(a,b,10)
fig,axes=plt.subplots(1,2,figsize=(14,6))
axes[0].contour(X,Y,f(X,Y),15)
axes[0].fill_between(xint,g1(xint),h1(xint),color='grey',alpha=0.5)    # integral region
axes[0].set_title('Integral over shaded region: %e\nestimated error: %e'%(val1,err1))



axes[1].contour(X,Y,f(X,Y),15)
axes[1].fill_between(xint,g2(xint),h2(xint),color='grey',alpha=0.5)    # integral region
axes[1].set_title('Integral over shaded region: %e\nestimated error: %e'%(val2,err2))



plt.show()



```

We consider the easiest case: 
    $$
    \int_{0}^1\int_{0}^{1}\int_{0}^{1} e^{-x^2-y^2-z^2} dxdydz

```python
def f(x,y,z):
    return np.exp(-x**2-y**2-z**2)

a,b=0,1
g,h=lambda x:0, lambda x:1
q,r=lambda x,y:0, lambda x,y:1


integrate.tplquad(f,a,b,g,h,q,r)
```

We can also compute integrals of the following form:
$$
\displaystyle\int\dots\int_Df(\mathbf{x})d\mathbf{x}
$$

<font color=blue>**Example**:
    evaluate
    $$
    \int_{0}^1\dots\int_{0}^1 e^{-x_1^2-x_2^2-\dots -x_n^2} dx_1dx_2\dots dx_n 
    $$
</font>

How to use:

**integrate.nquad(f,[(a1,b1),(a_2,b2),....,(an,bn)])**

```python
def f(*args):  # arbitrary number of inputs: *args unpack the input args list
    return np.exp(-np.sum(np.array(args )**2))
```

```python
%time integrate.nquad(f,[(0,1)]*1)
```

```python
%time integrate.nquad(f,[(0,1)]*2)
```

```python
%time integrate.nquad(f,[(0,1)]*3)
```

```python
%time integrate.nquad(f,[(0,1)]*4)
```

```python
%time integrate.nquad(f,[(0,1)]*5)
```

We see that the complexity of a multiple integral grows quickly!
